```python
df1=pd.read_csv(r"D:\Data\kcc_dataset.csv")
q_types=['Credit','Training and Exposure Visits', 'Crop Insurance', 'Market Information', 'Government Schemes', 'Weather']
#state=['DELHI']
```

```python
# Filter the dataset
f_df = df1[(df1['Crop'].str.isnumeric() == False)]
        # #   & (df['KccAns'] != 'Thank you') & (df['KccAns'] != 'DETAILS GIVEN') &
        #   (df['QueryText'] != 'TEST CALL') & (df['QueryText'] != "Farmer asked query on Weather") &
        #   (df['QueryText'] != 'no voice') & (df['QueryText'] != "Blank Call") & 
        #   (df['QueryType'] != 'Weather') & (df['Crop'] != 'Others')]

df = pd.DataFrame(f_df)
#df= df[df['StateName'].isin(state)]
df=df[~df['QueryType'].isin(q_types)]
```

```python
cr=df.Crop.value_counts(normalize=False,sort=True,ascending=True)
cr_df=cr.reset_index()
cr_df.columns=['Crop','Query_count']
print(cr_df)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(5,2))
plt.bar(yr_df['Year'],yr_df['Query_count'], color ='Teal',width = 0.3)
plt.xlim([2009,2024])
#plt.title('Query count Plot',fontsize=15)
plt.xlabel('Year',fontsize=10)
plt.ylabel('Query count', fontsize=10)
plt.savefig(f"E:\Data\district_wise.png")
plt.show()
```

```python
df1['Place']= df1['DistrictName']+ '_' + df1['StateName']
dis=df1.StateName.value_counts(normalize=False,sort=True,ascending=True)
dis_df=dis.reset_index()
dis_df.columns=['DistrictName','Query_count']
print(dis_df)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(5,6))
plt.plot(dis_df.Query_count,dis_df.DistrictName,'DarkBlue')
plt.title('Query count plot',fontsize=10)
plt.ylabel('District',fontsize=10)
plt.xlabel('Query count', fontsize=10)
plt.show()
```
